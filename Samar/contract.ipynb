```python
current = 'CT'
# List of contracts
SB_contracts = [
    "SBV4", "SBH5", "SBK5", "SBN5",
    "SBV5", "SBH6", "SBK6", "SBN6",
    "SBV6", "SBH7", "SBK7", "SBN7"
]
CC_contracts = [
    "CCU4", "CCZ4", "CCH5", "CCK5", "CCN5",
    "CCU5", "CCZ5", "CCH6", "CCK6", "CCN6"
]

KC_contracts = [
    "KCU4", "KCZ4", "KCH5", "KCK5", "KCN5",
    "KCU5", "KCZ5", "KCH6", "KCK6", "KCN6",
    "KCU6", "KCZ6", "KCH7", "KCK7", "KCN7"
]
CT_contracts = [
    "CTV4", "CTZ4", "CTH5", "CTK5", "CTN5",
    "CTV5", "CTZ5", "CTH6", "CTK6", "CTN6",
    "CTV6", "CTZ6", "CTH7", "CTK7", "CTN7"
]

# Function to extract month and year from contract
def extract_month_year(contract):
    month = contract[2]  # The third character is the month code
    year = contract[3]   # The fourth character is the year code
    return month, year

# Function to generate butterfly spreads
def generate_butterflies(contracts):
    butterflies = []
    for i in range(len(contracts) - 2):
        for j in range(i + 1, len(contracts) - 1):
            for k in range(j + 1, len(contracts)):
                month1, year1 = extract_month_year(contracts[i])
                month2, year2 = extract_month_year(contracts[j])
                month3, year3 = extract_month_year(contracts[k])
                
                # Construct the fly prefix
                fly_prefix = f"{current}{month1}{month2}{month3}{year3}Fly"
                
                # Construct the full fly expression
                fly = f'{fly_prefix}("{contracts[i]}-{month2}{year2}"-"{contracts[j]}-{month3}{year3}")'
                butterflies.append(fly)
    return butterflies

# Generate all possible butterflies

butterflies = generate_butterflies(CT_contracts)

# Path to save the text file
txt_file_path = f'Structure/{current}_flies.txt'

# Save the double flies to a text file
with open(txt_file_path, 'w') as file:
    for fly in butterflies:
        file.write(fly + '\n')
# # Convert the list of butterflies to a DataFrame
df = pd.DataFrame(butterflies, columns=["Fly"])

# Load the existing Excel file
book = load_workbook(excel_file_path)

# Write the DataFrame to a new sheet in the existing Excel file
with pd.ExcelWriter(excel_file_path, engine='openpyxl') as writer:
    writer.book = book
    df.to_excel(writer, sheet_name=f'{current} Fly', index=False)

print(f"Fly spreads have been saved to the 'Butterfly Spreads' sheet in {excel_file_path}")
```

```python
current = 'CT'
# List of contracts
SB_contracts = [
    "SBV4", "SBH5", "SBK5", "SBN5",
    "SBV5", "SBH6", "SBK6", "SBN6",
    "SBV6", "SBH7", "SBK7", "SBN7"
]
CC_contracts = [
    "CCU4", "CCZ4", "CCH5", "CCK5", "CCN5",
    "CCU5", "CCZ5", "CCH6", "CCK6", "CCN6"
]

KC_contracts = [
    "KCU4", "KCZ4", "KCH5", "KCK5", "KCN5",
    "KCU5", "KCZ5", "KCH6", "KCK6", "KCN6",
    "KCU6", "KCZ6", "KCH7", "KCK7", "KCN7"
]
#"CTV4","CTV5","CTV6", 
CT_contracts = [
     "CTZ4", "CTH5", "CTK5", "CTN5",
     "CTZ5", "CTH6", "CTK6", "CTN6",
    "CTZ6", "CTH7", "CTK7", "CTN7"
]
# Function to extract month and year from contract
def extract_month_year(contract):
    month = contract[2]  # The third character is the month code
    year = contract[3]   # The fourth character is the year code
    return month, year

# Function to generate double fly spreads in the specified format
def generate_formatted_double_flies(contracts):
    double_flies = []
    for i in range(len(contracts) - 3):  # Ensure there are enough contracts for a double fly
        for j in range(i + 1, len(contracts) - 2):
            for k in range(j + 1, len(contracts) - 1):
                for l in range(k + 1, len(contracts)):
                    # Construct the fly prefix for the double fly
                    fly_prefix = f"{current}{extract_month_year(contracts[i])[0]}{extract_month_year(contracts[j])[0]}{extract_month_year(contracts[k])[0]}{extract_month_year(contracts[l])[0]}{extract_month_year(contracts[l])[1]}DoubleFly"

                    # Construct the full double fly expression
                    double_fly = f'{fly_prefix}("Q%{contracts[i]}"-("Q%{contracts[j]}"*3))+(("Q%{contracts[k]}"*3)-"Q%{contracts[l]}")'
                    double_flies.append(double_fly)
    return double_flies

# Generate all possible formatted double flies
double_flies = generate_formatted_double_flies(CT_contracts)

# Path to save the text file
txt_file_path = f'structure/{current}_double_flies.txt'

# Save the double flies to a text file
with open(txt_file_path, 'w') as file:
    for fly in double_flies:
        file.write(fly + '\n')

# Convert the list of double flies to a DataFrame
df = pd.DataFrame(double_flies, columns=["Double Fly"])

# Path to the existing Excel file
excel_file_path = 'structures.xlsx'

# Load the existing Excel file
book = load_workbook(excel_file_path)

# Write the DataFrame to a new sheet in the existing Excel file
with pd.ExcelWriter(excel_file_path, engine='openpyxl') as writer:
    writer.book = book
    df.to_excel(writer, sheet_name=f'{current} Double Fly', index=False)
print(f"Formatted double fly spreads have been saved to {txt_file_path}")
```
