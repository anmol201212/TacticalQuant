```python
import pandas as pd
import numpy as np
```

```python
def find_dynamic_range_bound_periods(df, threshold=1.5):
    range_bound_periods = []
    start_idx = 0

    while start_idx < len(df):
        max_high = df['High'].iloc[start_idx]
        min_low = df['Low'].iloc[start_idx]
        end_idx = start_idx

        # Expand the period as long as it stays within the threshold
        while end_idx < len(df):
            max_high = max(max_high, df['High'].iloc[end_idx])
            min_low = min(min_low, df['Low'].iloc[end_idx])
            
            # If price range exceeds threshold, stop expanding
            if (max_high - min_low) > threshold:
                break
            
            end_idx += 1

        # If a range-bound period of at least 2 days is found, store it
        if end_idx - start_idx > 1:
            range_bound_periods.append({
                'Start_Date': df['Date'].iloc[start_idx],
                'End_Date': df['Date'].iloc[end_idx - 1],
                'Price_Range': f"{min_low} to {max_high}",
                'Days': end_idx - start_idx
            })

        # Move to the next period
        start_idx = end_idx

    return pd.DataFrame(range_bound_periods)
```

```python
df = pd.read_csv('/workspaces/Futures-First/Prediction/data/Corn/ZC Dec-Mar/2024.csv')
df
```

```python
range_bound_periods = find_dynamic_range_bound_periods(df, threshold=1.5)

# Show the result
print("Range-bound periods:")
print(range_bound_periods)
```
